### Target Variable Transformation

```python
severity_map = {1: "Severe", 2: "Serious", 3: "Slight"}
df['Accident_Severity'] = df['Accident_Severity'].map(severity_map)
```

```python
df["Accident_Severity"].value_counts()
```

```python
sns.countplot(x='Accident_Severity', data=df)
plt.title("Distribution of Accident Severity")
plt.show()
```

Most accidents are slight in severity, while severe accidents occur very rarely.
This indicates a strong class imbalance in the dataset.

```python
sns.countplot(y='Weather_Conditions', data=df)
plt.title("Accidents by Weather Conditions")
plt.show()
```

The majority of accidents occur during fine weather without high winds.
This suggests that traffic volume has a greater impact on accidents than adverse weather.

```python
sns.histplot(df['Hour'].dropna(), bins=24)
plt.xticks(range(0,24))
plt.title("Accidents by Hour of Day (24-Hour Distribution)")
plt.xlabel("Hour of Day (0–23)")
plt.ylabel("Number of Accidents")
plt.show()
```

The 24-hour distribution of UK road accidents shows that accidents occur throughout the day, with distinct peaks during morning (7–9 AM) and evening (4–7 PM) rush hours.Accident frequency is lowest during late night and early morning hours (12 AM–5 AM).

```python
sns.countplot(
    x='Accident_Severity',
    hue='Light_Conditions',
    data=df
)
plt.title("Light Conditions vs Accident Severity")
plt.show()
```

This indicates that while daylight has a higher number of accidents due to traffic volume, poor lighting conditions increase accident severity risk.

```python
urban_map = {
    1: "Urban",
    2: "Rural",
    3: "Unallocated"
}

df['Urban_or_Rural_Area'] = df['Urban_or_Rural_Area'].map(urban_map)
```

```python
sns.countplot(x='Urban_or_Rural_Area', data=df)
plt.title("Urban vs Rural Accidents")
plt.show()
```

Most accidents occur in urban areas, followed by rural areas.
This is expected due to higher traffic density and population in urban regions.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm',annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

This heatmap helps identify relationships between numerical variables.
